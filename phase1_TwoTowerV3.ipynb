```python
import pandas as pd

# Load movies datasets
movies = pd.read_csv("movies.csv")      # Contains movieId, title, genres
#ratings = pd.read_csv("ratings.csv")    # Contains userId, movieId, rating, timestamp
TMDB = pd.read_csv("tmdb_5000_movies.csv") #contains movies data
TMDB.drop(['genres'],axis=1,inplace=True)
links = pd.read_csv("links.csv")

# Merge MovieLens movies with links to get tmdbId
movies_links = pd.merge(movies, links, on='movieId')

# Merge with TMDB dataset using tmdbId
merged = pd.merge(movies_links, TMDB, left_on='tmdbId', right_on='id')

# Now 'merged' contains MovieLens movies enriched with TMDB metadata
merged.head()
```

```python
#drop unused columns
merged = merged.drop(columns=['imdbId', 'budget','homepage','keywords','original_title','title_y','production_companies',
 'production_countries',
 'release_date',
 'revenue',
 'runtime',
 'spoken_languages',
 'status',
 'tagline','vote_average',
 'vote_count'])
```

```python
#merge with credits
credits = pd.read_csv("tmdb_5000_credits.csv")
merged_credits = pd.merge(merged, credits, left_on='tmdbId',right_on='movie_id')


merged_credits.head()
```

```python
merged_credits.count()
```

```python
#adding director and cast columns
import ast
# Convert cast and crew columns
merged_credits['cast'] = merged_credits['cast'].apply(ast.literal_eval)
merged_credits['crew'] = merged_credits['crew'].apply(ast.literal_eval)

# Flattened top 3 cast names
def get_top_cast(cast_list, top_n=3):
    names = [member['name'] for member in cast_list[:top_n]]
    return "|".join(names)

merged_credits['top_cast'] = merged_credits['cast'].apply(lambda x: get_top_cast(x))

# Director name separately
def get_director(crew_list):
    for member in crew_list:
        if member.get('job') == 'Director':
            return member.get('name')
    return None

merged_credits['director'] = merged_credits['crew'].apply(lambda x: get_director(x))

# Final dataframe
print(merged_credits[['top_cast', 'director']].head())
```

```python
#Preparing users data
ratings = pd.read_csv("ratings.csv")
ratings = ratings.drop(columns=['timestamp'])
ratings = ratings.merge(movies[['movieId', 'genres']], on='movieId', how='left')
#this is to make genres column as a list
ratings['genres'] = ratings['genres'].str.split('|')
ratings.head()
```

```python
ratings.count()
```

```python
#synthesizing genres column based on high
from collections import Counter
# Filter high ratings
high_ratings = ratings[ratings['rating'] >= 4.0]
high_ratings.loc[:,'genres'] = high_ratings['genres'].apply(
    lambda x: x if isinstance(x, list) else []
)
# Get preferred genres
def get_preferred_genres(user_ratings):
    genre_counter = Counter()
    for genres in user_ratings:
        genre_counter.update(genres)
    top_genres = [genre for genre, _ in genre_counter.most_common(3)]  # Top 3
    return top_genres

users = high_ratings.groupby('userId')['genres'].apply(get_preferred_genres).reset_index()
#users.columns = ['userId', 'preferred_genres']
```

```python
users.head()
```

```python
#synthesizing birthdate , nationality
import random
from datetime import datetime, timedelta

# Sample lists
nationalities = ['USA', 'Canada', 'UK', 'India', 'Germany', 'France']
locations = {
    'USA': ['New York', 'Los Angeles', 'Chicago'],
    'Canada': ['Toronto', 'Vancouver', 'Montreal'],
    'UK': ['London', 'Manchester', 'Birmingham'],
    'India': ['Mumbai', 'Delhi', 'Bangalore'],
    'Germany': ['Berlin', 'Munich', 'Frankfurt'],
    'France': ['Paris', 'Lyon', 'Marseille']
}

# Synthesizing function
def synthesize_user_data(userId):
    birth_year = random.randint(1960, 2005)
    birth_month = random.randint(1, 12)
    birth_day = random.randint(1, 28)  # Simplify to avoid date issues
    birthdate = datetime(birth_year, birth_month, birth_day).strftime('%Y-%m-%d')

    nationality = random.choice(nationalities)
    location = random.choice(locations[nationality])

    return pd.Series([birthdate, nationality, location])

# Apply for each user
users[['birthdate', 'nationality', 'location']] = users['userId'].apply(synthesize_user_data)
```

```python
# Import datetime if not already imported
from datetime import datetime

# Calculate age from birthdate
def calculate_age(birthdate):
    today = datetime.today()
    birthdate = datetime.strptime(birthdate, '%Y-%m-%d')
    return today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))

# Categorize age
def categorize_age(age):
    if age < 13:
        return 'Child'
    elif age < 18:
        return 'Teenager'
    elif age < 30:
        return 'Young'
    elif age < 50:
        return 'Mid Age'
    elif age < 65:
        return 'Old'
    elif age < 101:
        return 'Senior'
    else:
        return 'Unknown'

# Apply to dataframe
users['age'] = users['birthdate'].apply(calculate_age)
users['age_category'] = users['age'].apply(categorize_age)

# Preview
print(users[['userId', 'birthdate', 'age', 'age_category']].head())
```
